```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

x_test = pd.read_csv("./yemoonsaBigdata/datasets/Part2/mpg_X_test.csv")
x_train = pd.read_csv("./yemoonsaBigdata/datasets/Part2/mpg_X_train.csv")
y_train = pd.read_csv("./yemoonsaBigdata/datasets/Part2/mpg_y_train.csv")
```

```python
x_tr, x_val, y_tr, y_val = train_test_split(x_train, y_train, test_size=0.2, random_state=0)
```

```python
print(x_tr.isna().sum())
print(x_val.isna().sum())
```

```python
x_tr.dropna(axis = 1, inplace = True)
x_val.dropna(axis = 1, inplace = True)

print(x_tr.info())
print(x_val.info())
```

```python
cols = x_tr.select_dtypes(include = object)

le = LabelEncoder()

for col in cols:
  x_tr[col] = le.fit_transform(x_tr[col])

cols = x_val.select_dtypes(include = object)

le = LabelEncoder()

for col in cols:
  x_val[col] = le.fit_transform(x_val[col])
```

```python
min_max_scaler = MinMaxScaler()

x_tr_data = min_max_scaler.fit_transform(x_tr)
x_val_data = min_max_scaler.fit_transform(x_val)

x_tr = pd.DataFrame(x_tr_data, columns = x_tr.columns)
x_val = pd.DataFrame(x_val_data, columns = x_val.columns)
```

```python
y_val
```

```python
dtc = DecisionTreeClassifier()
model = dtc.fit(x_tr, y_tr)
predict = model.predict(x_val)
```

```python
print(accuracy_score(y_val['isUSA'], predict))
```

```python
submit = y_val
submit['isUSA'] = predict
submit.head(5)
```

```python
submit.to_csv('submit.csv', index=False)
```
