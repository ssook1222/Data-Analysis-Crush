```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import LabelEncoder

from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.tree import DecisionTreeRegressor

x_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/carsprice/X_train.csv")
y_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/carsprice/y_train.csv")
x_test= pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/carsprice/X_test.csv")
y_test = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/carsprice/y_test.csv")

print(x_train.isna().sum())
print(x_test.isna().sum())
```

```python
print(x_train.info())
print(x_test.info())
```

```python
x_train.drop(columns = ['carID'], axis = 1, inplace = True)
x_test.drop(columns = ['carID'], axis = 1, inplace = True)
```

```python
le = LabelEncoder()

cols = x_train.select_dtypes(include = object)

for col in cols:
  x_train[col] = le.fit_transform(x_train[col])
  x_test[col] = le.transform(x_test[col])
```

```python
min_max_scaler = MinMaxScaler()
x_train_data = min_max_scaler.fit_transform(x_train)
x_test_data = min_max_scaler.fit_transform(x_test)

x_train = pd.DataFrame(x_train_data, columns = x_train.columns)
x_test = pd.DataFrame(x_test_data, columns = x_test.columns)
```

```python
dtr = DecisionTreeRegressor()
model = dtr.fit(x_train, y_train)
predict = model.predict(x_test)
```

```python
print(mean_absolute_error(y_test['price'], predict[:,1]))
```
