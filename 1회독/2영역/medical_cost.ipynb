```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import LabelEncoder

from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error
```

```python
x_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/MedicalCost/x_train.csv")
y_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/MedicalCost/y_train.csv")
x_test = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/MedicalCost/x_test.csv")
y_test = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/MedicalCost/y_test.csv")
```

```python
print(x_train.isna().sum())
print(x_test.isna().sum())
```

```python
print(x_train.info())
print(x_test.info())
```

```python
cols = x_train.select_dtypes(include = object)

le = LabelEncoder()

for col in cols:
  x_train[col] = le.fit_transform(x_train[col])
  x_test[col] = le.transform(x_test[col])
```

```python
min_max = MinMaxScaler()

x_train_data = min_max.fit_transform(x_train)
x_test_data = min_max.transform(x_test)

x_train = pd.DataFrame(x_train_data, columns = x_train.columns)
x_test = pd.DataFrame(x_test_data, columns = x_test.columns)
```

```python
dtr = DecisionTreeRegressor(random_state = 42)
model = dtr.fit(x_train, y_train)
predict = model.predict(x_test)
```

```python
y_test
```

```python
print(mean_absolute_error(y_test['charges'], predict[:, 1]))
```
