### Question 57

Income_Category의 카테고리를 map 함수를 이용하여 다음과 같이 변경하여 newIncome 컬럼에 매핑하라

```python
df_copy = df
```

```python
def check(x):
  if x == 'Unknown':
    return 'N'

  elif x == 'Less than $40K':
    return 'a'

  elif x == '$40K - $60K':
    return 'b'

  elif x == '$60K - $80K':
    return 'c'

  elif x == '$80K - $120K':
    return 'd'

  elif x == '$120K +':
    return 'e'

df['newIncome'] = df_copy.Income_Category.map(check)
df
```

### Question 58

Income_Category의 카테고리를 apply 함수를 이용하여 다음과 같이 변경하여
newIncome 컬럼에 매핑하라     
Unknown : N
Less than $40K : a
$40K - $60K : b
$60K - $80K : c
$80K - $120K : d
$120K +’ : e

```python
def check(x):
  if x == 'Unknown':
    return 'N'

  elif x == 'Less than $40K':
    return 'a'

  elif x == '$40K - $60K':
    return 'b'

  elif x == '$60K - $80K':
    return 'c'

  elif x == '$80K - $120K':
    return 'd'

  elif x == '$120K +':
    return 'e'

df['newIncome'] = df_copy.Income_Category.apply(check)
df
```

### Question 59

Customer_Age의 값을 이용하여 나이 구간을 AgeState 컬럼으로 정의하라.    
(0 ~ 9 : 0, 10 ~ 19 :10 , 20 ~ 29 :20 … 각 구간의 빈도수를 출력하라

```python
def check(x):
  if 0 <= x <= 9 :
    return 0

  elif 10 <= x <= 19:
    return 10

  elif 20 <= x <= 29:
    return 20

  elif 30 <= x <= 39:
    return 30

  elif 40 <= x <= 49:
    return 40

  elif 50 <= x <= 59:
    return 50

  elif 60 <= x <= 69:
    return 60

  elif 70 <= x <= 79:
    return 70


df['AgeState'] = df_copy.Customer_Age.apply(check)
df.AgeState.value_counts().sort_index(ascending = True)
```

### Question 60

Education_Level의 값중 Graduate단어가 포함되는 값은 1 그렇지 않은 경우에는 0으로 변경하여 newEduLevel 컬럼을 정의하고 빈도수를 출력하라

```python
def check(x):
  if 'Graduate' in x:
    return 1
  else:
    return 0

df['newEduLevel'] = df.Education_Level.apply(check)
df.newEduLevel.value_counts()
```
