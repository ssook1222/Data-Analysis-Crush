## 03. Grouping

### Question 44

데이터를 로드하고 상위 5개 컬럼을 출력하라

```python
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/AB_NYC_2019.csv')
df.head(5)
```

### Question 45

데이터의 각 host_name의 빈도수를 구하고 host_name으로 정렬하여 상위 5개를 출력하라.

```python
df45 = pd.DataFrame(df.host_name.value_counts()).reset_index(drop = False)
df45.sort_values('host_name').head(5)
```

### Question 46

데이터의 각 host_name의 빈도수를 구하고 빈도수 기준 내림차순 정렬한 데이터 프레임을 만들어라. 빈도수 컬럼은 counts로 명명하라

```python
df46 = df45.sort_values('count', ascending = False)
df46.rename(columns={'count': 'counts'}, inplace=True)
df46
```

### Question 47

neighbourhood_group의 값에 따른 neighbourhood 컬럼 값의 갯수를 구하여라

```python
df.groupby(['neighbourhood_group', 'neighbourhood'],as_index = False).size()
```

### Question 48

neighbourhood_group의 값에 따른 neighbourhood컬럼 값 중 neighbourhood_group그룹의 최댓값들을 출력하라

```python
df48 = df.groupby(['neighbourhood_group','neighbourhood'], as_index=False).size().groupby(['neighbourhood_group'], as_index=False).max()
df48
```

### Question 49

neighbourhood_group 값에 따른 price값의 평균, 분산, 최대, 최소 값을 구하여라

```python
df.groupby('neighbourhood_group', as_index = False).price.agg(['mean', 'var', 'min', 'max'])
```

### Question 50

neighbourhood_group 값에 따른 reviews_per_month 평균, 분산, 최대, 최소 값을 구하여라

```python
df.groupby('neighbourhood_group', as_index = False).reviews_per_month.agg(['mean', 'var', 'min', 'max'])
```
