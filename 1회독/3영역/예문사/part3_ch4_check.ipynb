```python
import pandas as pd
from scipy import stats
import math
```

```python
df = pd.read_csv('./yemoonsaBigData/datasets/supplement/airquality.csv')
```

### AirQuality

1-1. 표본평균 x를 구하시오.

```python
wind_mean = df['Wind'].mean()
print(round(wind_mean, 2))
```

1-2. 위의 가설을 검정하기 위한 검정 통계량을 구하시오.

```python
raw_ans = stats.ttest_1samp(df['Wind'], 10)
print(raw_ans)
ans = raw_ans.statistic

print(round(ans, 2))
```

1-3. pvalue를 구하고 가설 검정 결과

```python
print(round(raw_ans.pvalue, 2))
```

3-2. 검정통계량

```python
x1 = df3[(df3['인종'] == 'A그룹')]['수학 점수']
x2 = df3[(df3['인종'] == 'B그룹')]['수학 점수']
x3 = df3[(df3['인종'] == 'C그룹')]['수학 점수']
x4 = df3[(df3['인종'] == 'D그룹')]['수학 점수']


print(stats.f_oneway(x1, x2, x3, x4))
raw_ans = stats.f_oneway(x1, x2, x3, x4)

print(round(raw_ans.statistic, 2))
```

3-3. 가설 검정

```python
print(round(raw_ans.pvalue, 2))
```

### StudentsPerformance2

```python
df4 = pd.read_csv('./yemoonsaBigData/datasets/supplement/StudentsPerformance2.csv',  encoding = 'EUC-KR')
```

```python
df4
```

4-1. 학업 성적이 우수한 남학생 수와 여학생 수 차를 정수로 출력

```python
female = df4[((df4['학업 성적'] == '우수') & (df4.성별 == '여성'))].count().values[0]
male = df4[((df4['학업 성적'] == '우수') & (df4.성별 == '남성'))].count().values[0]

print(abs(male - female))
```

4-2. 검정통계량

```python
cross_tab = pd.crosstab(df4['학업 성적'], df4['성별'])
```

```python
cross_tab
```

```python
x1 = cross_tab.loc[:, '남성']
x2 = cross_tab.loc[:, '여성']
```

```python
raw_ans = stats.chi2_contingency([x1,x2]) # 리스트로 묶어주는 게 필수!!
print(raw_ans)
print(round(raw_ans.statistic, 2))
```

4-3. pvalue를 구하여라

```python
print(round(raw_ans.pvalue, 2))
```
