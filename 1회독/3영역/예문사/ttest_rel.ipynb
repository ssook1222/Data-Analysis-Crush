```python
import pandas as pd

df = pd.read_csv('./yemoonsaBigData/datasets/supplement/insectsprays.csv')
```

```python
from scipy import stats
import math
```

1. 표본평균을 구하시오.

```python
before = df['before_spr']
after = df['after_spr']

ans = (before - after).mean()
print(round(ans, 2))
```

2. 검정 통계량을 구하시오.

```python
raw_ans = stats.ttest_rel(before, after, alternative = 'greater') # alternative로 아무 값을 입력 안하면 default로 two-side. 문제에서는 greater
print(raw_ans)
print(round(raw_ans.statistic, 2))
```

3. p value를 구하시오.

```python
print(round(raw_ans.pvalue, 2))
```

0.05보다 작으므로 대립가설 채택
