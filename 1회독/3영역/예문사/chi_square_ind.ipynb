```python
import pandas as pd
from scipy import stats
import math
```

```python
df = pd.read_csv('./yemoonsaBigData/datasets/supplement/student.csv')
```

1. 분할표를 만들어라.

```python
df
```

```python
ans1 = pd.crosstab(df['단과대학'], df['수강 과목 수'])
print('경영대학')
```

2. 검정 통계량을 구하시오.

```python
x1 = ans1.loc['경영대학']
x2 = ans1.loc['공과대학']
x3 = ans1.loc['자연대학']
```

```python
raw_ans = stats.chi2_contingency([x1, x2, x3])
print(raw_ans)
print(round(raw_ans.statistic, 2))
```

3. pvalue를 구하시오.

```python
print(round(raw_ans.pvalue, 2))
```
