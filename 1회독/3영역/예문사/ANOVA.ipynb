```python
import pandas as pd
from scipy import stats
import math
```

```python
df = pd.read_csv('./yemoonsaBigData/datasets/supplement/iris.csv')
```

1. 특정 종의 평균 꽃받침 너비를 구하시오

```python
versi_mean = df[(df.species == 'versicolor')].sepal_width.mean()
print(round(versi_mean, 2))
```

2. 위의 가설을 검정하기 위한 검정 통계량을 구하시오.

```python
x1 = df[(df.species == 'versicolor')].sepal_width
x2 = df[(df.species == 'setosa')].sepal_width
x3 = df[(df.species == 'virginica')].sepal_width
```

```python
print(stats.f_oneway(x1, x2, x3))

raw_ans = stats.f_oneway(x1, x2, x3)
print(round(raw_ans.statistic, 2))
```

3. pvalue를 구하고 대립가설 기각/채택 구하기

```python
print(raw_ans.pvalue)
```

0.05보다 훨씬 작은 값이므로 유의, 대립가설 채택.
