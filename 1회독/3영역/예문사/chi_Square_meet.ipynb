```python
import pandas as pd
from scipy import stats
import math
```

```python
df = pd.read_csv('./yemoonsaBigData/datasets/supplement/cellphone.csv')
```

1. 삼성 - 애플 정수로 출력

```python
samsung = df[(df.제조사 == '삼성')].count().values[0]
apple = df[(df.제조사 == '애플')].count().values[0]
other = df[(df.제조사 == '기타')].count().values[0]

diff = samsung - apple
print(diff)
```

2. 기대빈도 계산하고 검정통계량

```python
total = df.count().values[0]

expect_samsung = total * 0.6
expect_apple = total * 0.3
expect_other = total * 0.1

expect = [expect_samsung, expect_apple, expect_other]
```

```python
see = [samsung, apple, other]
```

```python
print(stats.chisquare(see, f_exp = expect))

raw_ans = stats.chisquare(see, f_exp = expect)
print(round(raw_ans.statistic, 2))
```

3. pvalue 출력 및 가설 검증 결과

```python
print(round(raw_ans.pvalue, 2))
```
