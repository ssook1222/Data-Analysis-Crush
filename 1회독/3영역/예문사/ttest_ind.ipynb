```python
import pandas as pd
import math
from scipy import stats
```

```python
df = pd.read_csv('./yemoonsaBigData/datasets/supplement/toothgrowth.csv')
```

1. 표본평균 각각을 구하시오.

```python
means_oj = df[(df.supp == 'OJ')].len.mean()
means_vc = df[(df.supp == 'VC')].len.mean()

print(means_oj, means_vc)
```

2. 검정 통계량을 구하시오.

```python
raw = stats.ttest_ind(df[(df.supp == 'OJ')].len, df[(df.supp == 'VC')].len, equal_var = True)
print(raw.statistic)
```

3. p value를 구하고 유의수준 0.05 하에서 가설 검정의 결과를 채택/기각 중 하나로 선택하시오.

```python
print(raw.pvalue)
```

p-value가 0.05보다 크므로 영가설 채택. 오렌지주스나 아스코르브산 투여 상관 없이 동일함.
