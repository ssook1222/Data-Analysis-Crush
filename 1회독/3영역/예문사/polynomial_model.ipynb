```python
import pandas as pd
import numpy as np
```

```python
# 데이터 생성
np.random.seed(0)
house_size = np.random.uniform(50, 250, 100)  # 주택 크기: 50제곱미터에서 250제곱미터
house_price = 5000 + 20 * house_size + 0.2 * house_size**2 + np.random.normal(0, 2000, 100)  # 주택 가격: 강한 비선형 관계

# DataFrame 생성
df = pd.DataFrame({
    'Size': house_size,
    'Price': house_price
})

# 다항 특성 생성: Size^2
df['Size^2'] = df['Size']**2
```

```python
import statsmodels.api as sm
```

```python
df
```

```python
x = df[['Size', 'Size^2']]
y = df['Price']
```

```python
model = sm.OLS(y, x).fit()
```

```python
print(model.summary())
```
