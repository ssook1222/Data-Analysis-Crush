### 2영역

```python
import pandas as pd
train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e7_p2_train2.csv')
test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e7_p2_test2.csv')
```

```python
import numpy as np

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler

from sklearn.metrics import mean_absolute_error
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
```

```python
train.columns
```

```python
x = train[['연월', '업종명', '이용자구분', '성별', '이용자수', '이용건수']]
y = train[['이용금액']]
```

```python
x_tr, x_val, y_tr, y_val = train_test_split(x, y, test_size = 0.3)
```

```python
print(x_tr.isna().sum())
print(x_val.isna().sum())

print(x_tr.info())

cols = x_tr.select_dtypes(include = object)

le = LabelEncoder()

for col in cols:
  x_tr[col] = le.fit_transform(x_tr[col])
  x_val[col] = le.fit_transform(x_val[col])
```

```python
dtr = DecisionTreeRegressor()
model = dtr.fit(x_tr, y_tr)
predict_raw = dtr.predict(x_val)

print(mean_absolute_error(predict_raw, y_val) ** 0.5)
```

```python
xr = test[['연월', '업종명', '이용자구분', '성별', '이용자수', '이용건수']]
```

```python
print(xr.isna().sum())
for col in cols:
  xr[col] = le.fit_transform(xr[col])
```

```python
predict = dtr.predict(xr)
```

```python
submit = pd.DataFrame(predict, columns = y_tr.columns)

print(submit)
```

```python
submit.to_csv('./result.csv', index = False)
```
