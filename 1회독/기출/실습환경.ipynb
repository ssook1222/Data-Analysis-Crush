### 2영역

```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler

from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split

train = pd.read_csv("data/customer_train.csv")
test = pd.read_csv("data/customer_test.csv")

x = train[[ '총구매액', '최대구매액', '환불금액', '주구매상품', '주구매지점', '방문일수', '방문당구매건수','주말방문비율', '구매주기']]
y = train[['성별']]
x_tr, x_val, y_tr, y_val = train_test_split(x, y, test_size = 0.3)

print(x_tr.isna().sum())
x_tr = x_tr.fillna(0)
x_val = x_val.fillna(0)
test = test.fillna(0)
test = test[[ '총구매액', '최대구매액', '환불금액', '주구매상품', '주구매지점', '방문일수', '방문당구매건수',
       '주말방문비율', '구매주기']]
cols_x = x_tr.select_dtypes(include = object)
cols_y = y_tr.select_dtypes(include = object)

le = LabelEncoder()
min_max = MinMaxScaler()

for col in cols_x:
	x_tr[col] = le.fit_transform(x_tr[col])
	x_val[col] = le.fit_transform(x_val[col])
	test[col] = le.fit_transform(test[col])

for col in cols_y:
	y_tr[col] = le.fit_transform(y_tr[col])

dtc = DecisionTreeClassifier()
model = dtc.fit(x_tr, y_tr)
predict = model.predict(x_val)

print(accuracy_score(predict, y_val))
predict = model.predict(test)
submit = pd.DataFrame(predict, columns = y_val.columns)
submit['pred'] = submit['성별']
ans = submit['pred']
print(ans)
ans.to_csv('./result.csv', index = False)
```
