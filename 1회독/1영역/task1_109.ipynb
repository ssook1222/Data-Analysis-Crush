## nba 선수 능력치 데이터

```python
import pandas as pd
pd.set_option('display.max_columns',50)
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/nba/nba.csv",encoding='latin',sep=';')
df.head()
```

### Question 104

Tm 컬럼은 각 팀의 이름을 의미한다. TOR팀의 평균나이를 소수 4째 자리까지 구하여라

```python
round(df[(df.Tm == 'TOR')].Age.mean(), 4)
```

### Question 105

Pos 컬럼은 포지션을 의미한다. 전체 선수 중 최소나이대의 선수들을 필터하고 그들 중 가장 많은 포지션은 무엇인지 확인하라

```python
df[(df.Age == df.Age.min())].Pos.value_counts().index[0]
```

### Question 106

선수들의 이름은 first_name+ 공백 + last_name으로 이루어져 있다. 가장 많은 first_name은 무엇이며 몇 회 발생하는지 확인하라

```python
df['first_name'] = [name.split()[0] for name in df['Player']]
df.first_name.value_counts().head(1)
```

### Question 107

PTS컬럼은 경기당 평균득점수 이다. 각포지션별로 경기당 평균득점수의 평균을 구하여라

```python
df.groupby('Pos', as_index = False).PTS.mean()
```

### Question 108

- 107번과 문제 동일            

PTS컬럼은 경기당 평균득점수 이다. 각포지션별로 경기당 평균득점수의 평균을 구하여라

```python
df.groupby('Pos', as_index = False).PTS.mean()
```

### Question 109

G컬럼은 참여한 경기의 숫자이다. 각 팀별로 가장 높은 경기참여 수를 가진 선수들의 경기 참여 숫자의 평균을 구하여라

```python
df.groupby('Tm', as_index = False).G.max().G.mean()
```

### Question 110

Tm의 값이 MIA이며 Pos는 C또는 PF인 선수의 MP값의 평균은?

```python
df[(df.Tm == 'MIA') & ((df.Pos == 'C') | (df.Pos == 'PF'))].MP.mean()
```

### Question 111

전체 데이터중 G의 평균값의 1.5배 이상인 데이터들만 추출했을때 3P값의 평균은?

```python
df[(df.G >= df.G.mean() * 1.5)]['3P'].mean()
```

### Question 112

Age의 평균 이상인 그룹과 평균 미만인 그룹간의 G값의 평균의 차이는?

```python
df[(df.Age >= df.Age.mean())].G.mean() - df[(df.Age < df.Age.mean())].G.mean()
```

### Question 113

평균나이가 가장 젊은 팀은 어디인가?

```python
df.groupby('Tm', as_index = False).Age.mean().sort_values('Age').head(1).Tm.values[0]
```

### Question 114
