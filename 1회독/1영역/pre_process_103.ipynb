```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/chipo.csv')
df
```

### Question 22.
quantity컬럼 값이 3인 데이터를 추출하여 index를 0부터 정렬하고 첫 5행을 출력하라

```python
df[(df['quantity'] == 3)].reset_index(drop = True).head(5)
```

### Question 23
quantity , item_price 두개의 컬럼으로 구성된 새로운 데이터 프레임을 정의하라

```python
df23 = df[['quantity', 'item_price']]
df23
```

### [!] Question 24.
item_price 컬럼의 달러표시 문자를 제거하고 float 타입으로 저장하여 new_price 컬럼에 저장하라

```python
df23['new_price'] = df23.item_price.str[1:].astype('float')
df['new_price'] = df23.item_price.str[1:].astype('float')
df23
```

### Question 25.
new_price 컬럼이 5이하의 값을 가지는 데이터프레임을 추출하고, 전체 갯수를 구하여라



```python
df23[(df23.new_price <= 5)].count().values[0]
```

### Question 26.
item_name명이 Chicken Salad Bowl 인 데이터 프레임을 추출하라고 index 값을 초기화 하여라

```python
df26 = df[(df.item_name == 'Chicken Salad Bowl')].reset_index(drop = True)
df26
```

### Question 27.
new_price값이 9 이하이고 item_name 값이 Chicken Salad Bowl 인 데이터 프레임을 추출하라

```python
df27 = df[(df.new_price <= 9) & (df.item_name == 'Chicken Salad Bowl')]
df27
```

### Question 28.
df의 new_price 컬럼 값에 따라 오름차순으로 정리하고 index를 초기화 하여라

```python
df28 = df.sort_values('new_price').reset_index(drop = True)
df28
```

### Question 29.
df의 item_name 컬럼 값중 Chips 포함하는 경우의 데이터를 출력하라

```python
df29 = df[(df.item_name.str.contains('Chips'))]
df29
```

### Question 30.
df의 짝수번째 컬럼만을 포함하는 데이터프레임을 출력하라

```python
df30 = df[(df.index % 2 == 0)]
df30
```
