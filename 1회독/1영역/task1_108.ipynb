### Question 92

### Question 98

남성들의 연령대별 (10살씩 구분 0 ~ 9세 10 ~ 19세 …) Na_to_K값의 평균값을 구해서 데이터 프레임으로 표현하여라

```python
def age(x):
  if 0 <= x <= 9:
    return '0대'

  if 10 <= x <= 19:
    return '10대'

  if 20 <= x <= 29:
    return '20대'

  if 30 <= x <= 39:
    return '30대'

  if 40 <= x <= 49:
    return '40대'

  if 50 <= x <= 59:
    return '50대'

  if 60 <= x <= 69:
    return '60대'

  if 70 <= x <= 79:
    return '70대'

df['age_time'] = df.Age.apply(age)
df[['age_time', 'Na_to_K']].groupby('age_time', as_index = False).mean()
```
