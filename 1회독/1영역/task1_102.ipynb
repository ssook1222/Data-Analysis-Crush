### Question 13.

### Question 14.
각 채널의 2021-10-03 03:00:00 ~ 2021-11-01 15:00:00 까지 구독자수 (subcnt) 의 증가량을 구하여라

```python
first = channel[(channel.ct >= '2021-10-03 03:00:00')].drop_duplicates('channelid', keep = 'first')[['channelname', 'subcnt']].reset_index(drop = True)
second = channel[(channel.ct <= '2021-11-01 15:00:00')].drop_duplicates('channelid', keep = 'last')[['channelname', 'subcnt']].reset_index(drop = True)

second['increase'] = second.subcnt - first.subcnt

print(second[['channelname', 'increase']])
```

### Question 17.
“공범” 컨텐츠의 경우 19:00시에 공개 되는것으로 알려져있다.    
공개된 날의 21시의 viewcnt, ct, videoname 으로 구성된 데이터 프레임을 viewcnt를 내림차순으로 정렬하여 출력하라

```python
video['time'] = pd.to_datetime(video['ct']).dt.hour

df17 = video[(video.time == 21)].sort_values(['videoname', 'ct']).drop_duplicates('videoname').sort_values('viewcnt', ascending = False)[['videoname', 'viewcnt', 'ct']]
df17 = df17.reset_index(drop = True)

print(df17)
```

### Question 18.
video 정보의 가장 최근 데이터들에서 각 에피소드의 싫어요/좋아요 비율을   
ratio 컬럼으로 만들고 videoname, ratio로 구성된 데이터 프레임을 ratio를 오름차순으로 정렬하라.

```python
df18 = video.sort_values('ct').drop_duplicates('videoname', keep = 'last')
df18['ratio'] = df18.dislikecnt / df18.likecnt
print(df18[['videoname', 'ratio']].sort_values('ratio'))
```

### Question 19.
2021-11-01 00:00:00 ~ 15:00:00까지 각 에피소드별 viewcnt의 증가량을 데이터 프레임으로 만드시오

```python
video['ct'] = pd.to_datetime(video.ct)
```

```python
start = pd.to_datetime("2021-11-01 00:00:00")
end = pd.to_datetime("2021-11-01 15:00:00")

target = video[(video.ct >= start) & (video.ct <= end)].reset_index(drop=True)

def check(x):
    result = max(x) - min(x)
    return result

answer = target[['videoname','viewcnt']].groupby("videoname", as_index = False).agg(check)
answer
```
